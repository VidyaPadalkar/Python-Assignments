```python
import yfinance as yf

tickers = ['AAPL', 'GOOGL', 'META']
start_date = '2013-08-14'
end_date = '2023-08-14'

dfs = {}
for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)['Adj Close']
    dfs[ticker] = data

apple_df = pd.DataFrame(dfs['AAPL'])
google_df = pd.DataFrame(dfs['GOOGL'])
meta_df = pd.DataFrame(dfs['META'])
```

```python
merged_df = pd.concat([apple_df,google_df, meta_df],axis=1,keys=['Apple','Google','Meta'])
merged_df
```

```python
import numpy as np
log_returns = merged_df.apply(lambda x: np.log(x / x.shift(1)))
log_returns
```
